## 1.0 Objetivo:

As doenças cardiovasculares são consideradas as principais causas de morte no mundo e muitas vezes provocadas por maus hábitos da população, a má alimentação, tabagismo e o consumo de bebida alcoólica podem ser considerados exemplos. Visto isto, este presente trabalho tem como objetivo classificar pacientes com Doenças cardíacas, utilizando o conceito de redes neurais. Neste, será utilizado o Perceptron, uma rede neural artificial cujo pesos e inclinações podem ser treinados para produzir um vetor alvo que quando apresentamos tem que corresponder ao vetor de entrada.


## 2.0 Introdução:

 <p> As doenças cardiovasculares são um conjunto de doenças que atacam o coração e/ou os vasos sanguíneos. Dentre elas, podemos citar o enfarte agudo do miocárdio, arritmia cardíaca e os acidentes vasculares cerebrais (AVC).[1]</p> 
 <p>   Em 2014, a organização mundial da saúde OMS considerou as doenças cardíacas como principais causas de morte do mundo, estima-se que no Brasil morrem anualmente 300 mil pessoas em decorrência destas doenças.[2]</p> 
 <p>  Segundo o cardiologista Leonardo Spencer, do Hospital do Coração do Brasil, em Brasília. Além dos fatores genéticos, as causas das doenças cardiovasculares estão ligadas aos maus hábitos da população, como sedentarismo, sobrepeso, tabagismo. [2]</p> 
  <p> Estas condições, consideradas fatores de risco, aumenta consideravelmente as chances de ocorrência do problema cardíaco. De um modo geral, a hipertensão o sedentarismo, diabetes, hiperlipidemia, o tabagismo o álcool assim como a idade são os principais fatores de risco que colaboram com a ocorrência das doenças cardíacas.</p> 
<p>   Dado a gravidade de tais doenças a prevenção continua sendo uma alternativa altamente viável. Sendo assim, o diagnóstico precoce pode ser determinante no tratamento e na cura do paciente. Uma das alternativas, é o uso da tecnologia como aliada no diagnóstico precoce destas doenças, dentre elas, as redes neurais.  </p> 
   <p>  Uma rede neural artificial é uma ferramenta computacional, inspirada no cérebro humano capaz de realizar processamento de sinais, imagens e outros. Sua precisão e versatilidade permitem que essa ferramenta tenha diversas aplicações, na medicina, biologia, engenharia, matemática e outros. </p> 
	O perceptron, é um  tipo de rede neural artificial inspirado no neurônio. De modo geral, trata-se de um algoritmo de capaz de aprender e classificar atuando como um  classificador binário .  </p> 
    <p> Conforme apresentado na figura 1, sua arquitetura consiste na combinação linear das entradas e seus respectivos pesos, submetidas a uma junção somadora que ao ser subtraída pela função de ativação produz uma saída binária. </p>
